```python
import getpass

COPERNICUS_USERNAME = input("Enter Copernicus Marine username/email: ")
COPERNICUS_PASSWORD = getpass.getpass("Enter Copernicus Marine password: ")
```

```python
import copernicusmarine
from pathlib import Path

# ============================================================
# Copernicus Marine Global Ocean Waves Reanalysis
# Product: GLOBAL_MULTIYEAR_WAV_001_032
# Dataset: 3-hourly instantaneous wave data
# ============================================================

DATASET_ID = "cmems_mod_glo_wav_my_0.2deg_PT3H-i"

OUTPUT_DIR = Path("copernicusmarine_data/waves_reanalysis")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

BASINS = {
    "bob": {
        "name": "Bay_of_Bengal",
        "minimum_longitude": 80,
        "maximum_longitude": 98,
        "minimum_latitude": 5,
        "maximum_latitude": 22
    },
    "as": {
        "name": "Arabian_Sea",
        "minimum_longitude": 60,
        "maximum_longitude": 75,
        "minimum_latitude": 5,
        "maximum_latitude": 22
    }
}

YEARS = [2020, 2021, 2022, 2023, 2024]

WAVE_VARIABLES = ["VHM0", "VTM02", "VMDR"]


def download_wave_subset(basin_key, year):
    basin = BASINS[basin_key]

    output_filename = f"CMEMS_WAV_{basin_key}_VHM0_VTM02_VMDR_{year}.nc"

    print("\nDownloading wave data")
    print("Basin:", basin["name"])
    print("Year:", year)
    print("Output:", output_filename)

    copernicusmarine.subset(
        username=COPERNICUS_USERNAME,
        password=COPERNICUS_PASSWORD,
        dataset_id=DATASET_ID,
        variables=WAVE_VARIABLES,
        minimum_longitude=basin["minimum_longitude"],
        maximum_longitude=basin["maximum_longitude"],
        minimum_latitude=basin["minimum_latitude"],
        maximum_latitude=basin["maximum_latitude"],
        start_datetime=f"{year}-01-01T00:00:00",
        end_datetime=f"{year}-12-31T23:59:59",
        output_filename=output_filename,
        output_directory=str(OUTPUT_DIR),
        file_format="netcdf",
        netcdf_compression_level=4,
        skip_existing=True
    )


for basin_key in BASINS:
    for year in YEARS:
        download_wave_subset(basin_key, year)

print("All required Copernicus wave reanalysis files downloaded successfully.")
```

```python
import xarray as xr
import pandas as pd
import numpy as np
from pathlib import Path

INPUT_DIR = Path("copernicusmarine_data/waves_reanalysis")
OUTPUT_DIR = Path("copernicusmarine_data/waves_processed")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Wave energy approximation:
# P ≈ 0.49 × Hs² × T
# Hs = significant wave height, T = mean wave period
def compute_wave_energy(hs, period):
    return 0.49 * (hs ** 2) * period


def infer_basin_from_filename(filename):
    if "_bob_" in filename:
        return "Bay_of_Bengal"
    elif "_as_" in filename:
        return "Arabian_Sea"
    else:
        return "Unknown"


def process_one_wave_file(file_path):
    file_path = Path(file_path)
    print("Processing:", file_path.name)

    basin = infer_basin_from_filename(file_path.name)

    ds = xr.open_dataset(file_path)

    # Convert to dataframe
    df = ds[["VHM0", "VTM02", "VMDR"]].to_dataframe().reset_index()

    # Standardize column names
    df = df.rename(columns={
        "longitude": "lon",
        "latitude": "lat",
        "time": "timestamp",
        "VHM0": "significant_wave_height",
        "VTM02": "mean_wave_period",
        "VMDR": "mean_wave_direction"
    })

    df["basin"] = basin
    df["date"] = pd.to_datetime(df["timestamp"]).dt.date

    # Remove invalid values
    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.dropna(subset=["significant_wave_height", "mean_wave_period"])

    # Derived feature
    df["wave_energy_kw_per_m"] = compute_wave_energy(
        df["significant_wave_height"],
        df["mean_wave_period"]
    )

    # Daily grid-level aggregation
    daily = (
        df.groupby(["basin", "date", "lat", "lon"], as_index=False)
        .agg(
            wave_height_mean=("significant_wave_height", "mean"),
            wave_height_max=("significant_wave_height", "max"),
            wave_height_std=("significant_wave_height", "std"),
            wave_period_mean=("mean_wave_period", "mean"),
            wave_period_max=("mean_wave_period", "max"),
            wave_energy_mean=("wave_energy_kw_per_m", "mean"),
            wave_energy_max=("wave_energy_kw_per_m", "max"),
            wave_direction_mean=("mean_wave_direction", "mean"),
            valid_records=("significant_wave_height", "count")
        )
    )

    ds.close()
    return daily


all_files = sorted(INPUT_DIR.glob("*.nc"))
all_daily_frames = []

for file_path in all_files:
    try:
        daily_df = process_one_wave_file(file_path)
        all_daily_frames.append(daily_df)
    except Exception as e:
        print("Error:", file_path.name)
        print(e)

waves_daily = pd.concat(all_daily_frames, ignore_index=True)

print("Final shape:", waves_daily.shape)
waves_daily.head()
```

```python
import numpy as np
import pandas as pd
from pathlib import Path

OUTPUT_DIR = Path("copernicusmarine_data/waves_processed")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

df = waves_daily.copy()

# Convert date
df["date"] = pd.to_datetime(df["date"])

# Circular mean preparation for wave direction
df["wave_dir_sin"] = np.sin(np.deg2rad(df["wave_direction_mean"]))
df["wave_dir_cos"] = np.cos(np.deg2rad(df["wave_direction_mean"]))

waves_basin_daily = (
    df.groupby(["basin", "date"], as_index=False)
    .agg(
        wave_height_mean=("wave_height_mean", "mean"),
        wave_height_max=("wave_height_max", "max"),
        wave_height_std=("wave_height_mean", "std"),
        wave_period_mean=("wave_period_mean", "mean"),
        wave_period_max=("wave_period_max", "max"),
        wave_energy_mean=("wave_energy_mean", "mean"),
        wave_energy_max=("wave_energy_max", "max"),
        valid_grid_cells=("valid_records", "count"),
        total_valid_records=("valid_records", "sum"),
        wave_dir_sin_mean=("wave_dir_sin", "mean"),
        wave_dir_cos_mean=("wave_dir_cos", "mean")
    )
)

# Convert circular mean back to degrees
waves_basin_daily["wave_direction_mean"] = (
    np.rad2deg(
        np.arctan2(
            waves_basin_daily["wave_dir_sin_mean"],
            waves_basin_daily["wave_dir_cos_mean"]
        )
    ) + 360
) % 360

waves_basin_daily = waves_basin_daily.drop(
    columns=["wave_dir_sin_mean", "wave_dir_cos_mean"]
)

# Save compact summary
summary_csv = OUTPUT_DIR / "copernicus_waves_basin_daily_summary_2020_2024.csv"
summary_excel = OUTPUT_DIR / "copernicus_waves_basin_daily_summary_2020_2024.xlsx"
summary_parquet = OUTPUT_DIR / "copernicus_waves_basin_daily_summary_2020_2024.parquet"

waves_basin_daily.to_csv(summary_csv, index=False)
waves_basin_daily.to_excel(summary_excel, index=False)
waves_basin_daily.to_parquet(summary_parquet, index=False)

print("Summary rows:", len(waves_basin_daily))
print("Saved Excel summary:", summary_excel)

waves_basin_daily.head()
```
