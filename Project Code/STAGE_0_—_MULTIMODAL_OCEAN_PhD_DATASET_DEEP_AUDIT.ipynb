```python
# ============================================================
# STAGE 0 — MULTIMODAL OCEAN PhD DATASET DEEP AUDIT
# Colab 2026-compatible full analysis script
# Root path: /content/drive/MyDrive/Dataset/PHD
# ============================================================
# Purpose:
# 1. Analyse every dataset folder before model implementation.
# 2. Detect file counts, size, extension, validity, corrupted files, and schema.
# 3. Inspect NetCDF, GeoTIFF, CSV, Excel, and Parquet files safely.
# 4. Identify canonical ML-ready files for the later full PhD implementation.
# 5. Generate more than 10 figures and more than 8 tables for dataset readiness.
# 6. Save a complete Excel + Markdown + PNG figure report.
# ============================================================

# -----------------------------
# 0. Install and import packages
# -----------------------------
import sys
import subprocess
import importlib

REQUIRED_PACKAGES = [
    "pandas",
    "numpy",
    "openpyxl",
    "pyarrow",
    "xarray",
    "netCDF4",
    "h5py",
    "matplotlib",
    "tqdm",
    "rasterio",
]

for pkg in REQUIRED_PACKAGES:
    try:
        importlib.import_module(pkg if pkg != "pyarrow" else "pyarrow")
    except Exception:
        print(f"Installing missing package: {pkg}")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])

import os
import re
import json
import math
import mimetypes
import warnings
from pathlib import Path
from datetime import datetime
from collections import Counter, defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

warnings.filterwarnings("ignore")

try:
    from IPython.display import display
except Exception:
    display = print

# ------------------------------------------------------------
# Robust Google Drive mounting for Colab
# ------------------------------------------------------------
# Your previous error happened because /content/drive already contained
# local files, so Colab refused to mount Drive over that folder. This block
# checks whether /content/drive is a real mounted Drive. If it is not mounted
# and contains local files, it moves those local files to a backup folder,
# creates a clean mount point, and mounts Google Drive safely.
try:
    from google.colab import drive
    import subprocess
    import shutil

    DRIVE_MOUNT = Path("/content/drive")

    def is_real_mountpoint(path):
        try:
            return subprocess.run(
                ["mountpoint", "-q", str(path)],
                stdout=subprocess.DEVNULL,
                stderr=subprocess.DEVNULL
            ).returncode == 0
        except Exception:
            return False

    if not is_real_mountpoint(DRIVE_MOUNT):
        if DRIVE_MOUNT.exists() and any(DRIVE_MOUNT.iterdir()):
            backup_path = Path(f"/content/drive_local_backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}")
            print("/content/drive contains local files but is not mounted.")
            print("Moving local files to:", backup_path)
            shutil.move(str(DRIVE_MOUNT), str(backup_path))

        DRIVE_MOUNT.mkdir(parents=True, exist_ok=True)
        drive.mount(str(DRIVE_MOUNT), force_remount=True)
    else:
        print("Google Drive is already mounted at:", DRIVE_MOUNT)

except Exception as e:
    print("Google Drive mount failed or not running in Colab:", e)

# -----------------------------
# 1. Configuration
# -----------------------------
ROOT = Path("/content/drive/MyDrive/Dataset/PHD")
REPORT_DIR = ROOT / "_stage0_deep_dataset_audit"
FIG_DIR = REPORT_DIR / "figures"
TABLE_DIR = REPORT_DIR / "tables"

REPORT_DIR.mkdir(parents=True, exist_ok=True)
FIG_DIR.mkdir(parents=True, exist_ok=True)
TABLE_DIR.mkdir(parents=True, exist_ok=True)

# Large CSV files can be slow to count exactly. Keep this limit practical.
MAX_CSV_EXACT_ROWCOUNT_MB = 800
CSV_SAMPLE_ROWS = 10
PARQUET_BATCH_SIZE = 250_000

print("=" * 90)
print("STAGE 0: MULTIMODAL OCEAN PhD DATASET DEEP AUDIT")
print("=" * 90)
print("Dataset root:", ROOT)
print("Report folder:", REPORT_DIR)
print("Started at:", datetime.now())

if not ROOT.exists():
    raise FileNotFoundError(
        f"Dataset root not found: {ROOT}\n"
        "Please verify that Google Drive is mounted and the folder path is correct."
    )

# -----------------------------
# 2. Helper functions
# -----------------------------
def section(title):
    print("\n" + "=" * 90)
    print(title)
    print("=" * 90)


def short_line(text):
    print("-" * 90)
    print(text)
    print("-" * 90)


def human_size(num_bytes):
    if num_bytes is None or (isinstance(num_bytes, float) and np.isnan(num_bytes)):
        return "NA"
    try:
        num_bytes = float(num_bytes)
    except Exception:
        return "NA"
    if num_bytes <= 0:
        return "0 B"
    units = ["B", "KB", "MB", "GB", "TB"]
    idx = min(int(math.floor(math.log(num_bytes, 1024))), len(units) - 1)
    return f"{num_bytes / (1024 ** idx):.2f} {units[idx]}"


def top_folder(file_path):
    rel = file_path.relative_to(ROOT)
    return rel.parts[0] if len(rel.parts) > 1 else "__root__"


def safe_str(x, max_len=30000):
    if x is None:
        return ""
    text = str(x)
    text = text.replace("\x00", "")
    if len(text) > max_len:
        text = text[:max_len] + " ...[truncated]"
    return text


def read_signature(file_path, n=512):
    try:
        with open(file_path, "rb") as f:
            return f.read(n)
    except Exception:
        return b""


def detect_file_nature(file_path):
    sig = read_signature(file_path, 512)
    ext = file_path.suffix.lower()
    sig_l = sig.strip().lower()

    if len(sig) == 0:
        return "empty_or_unreadable"
    if sig.startswith(b"CDF"):
        return "valid_netcdf3"
    if sig.startswith(b"\x89HDF\r\n\x1a\n"):
        return "valid_netcdf4_hdf5"
    if sig.startswith(b"PK") and ext in [".xlsx", ".zip", ".docx", ".pptx"]:
        return "zip_container_or_office_file"
    if b"<html" in sig_l or b"<!doctype html" in sig_l:
        return "html_saved_as_data"
    if sig_l.startswith(b"{") or sig_l.startswith(b"["):
        return "json_or_api_response"
    if sig_l.startswith(b"<?xml"):
        return "xml_file"
    if ext in [".csv", ".txt", ".md"]:
        return "text_table_or_plain_text"
    if ext == ".parquet":
        return "parquet_file"
    if ext in [".tif", ".tiff"]:
        return "geotiff_or_raster"
    return "unknown_binary_or_other"


def safe_stat(file_path):
    try:
        st = file_path.stat()
        return int(st.st_size), datetime.fromtimestamp(st.st_mtime)
    except Exception:
        return None, None


def exact_csv_row_count(file_path):
    size_mb = file_path.stat().st_size / (1024 * 1024)
    if size_mb > MAX_CSV_EXACT_ROWCOUNT_MB:
        return np.nan, f"skipped_large_csv_over_{MAX_CSV_EXACT_ROWCOUNT_MB}MB"
    try:
        with open(file_path, "rb") as f:
            line_count = sum(1 for _ in f)
        return max(line_count - 1, 0), "exact"
    except Exception as e:
        return np.nan, f"failed:{e}"


def infer_component_from_path(path_text):
    p = path_text.lower()
    if "argo" in p and "bgc" not in p and "sprof" not in p:
        return "Argo_Core_Profile"
    if "bgc" in p or "sprof" in p:
        return "BGC_Argo"
    if "oisst" in p or "noaa_oisst" in p:
        return "NOAA_OISST"
    if "modis" in p or "chlorophyll" in p or "chlor_a" in p:
        return "MODIS_Chlorophyll"
    if "era5" in p:
        return "ERA5_Atmosphere"
    if "ibtracs" in p or "cyclone" in p:
        return "IBTrACS_Cyclone"
    if "gebco" in p or "bathymetry" in p:
        return "GEBCO_Bathymetry"
    if "rama" in p or "omni" in p or "incois" in p:
        return "INCOIS_OMNI_RAMA"
    if "wave" in p or "wav" in p:
        return "Copernicus_Waves"
    if "physics" in p or "current" in p or "uo_vo" in p or "mlotst" in p or "zos" in p:
        return "Copernicus_Physics"
    return "Unclassified"


def standardize_column_list(cols):
    return [str(c).strip() for c in cols]


def lower_join(values):
    return " ".join([str(v).lower() for v in values if pd.notna(v)])

# -----------------------------
# 3. File system inventory
# -----------------------------
section("1. Scanning every folder and file")

all_dirs = []
all_files = []
for p in ROOT.rglob("*"):
    if REPORT_DIR in p.parents or p == REPORT_DIR:
        continue
    if p.is_dir():
        all_dirs.append(p)
    elif p.is_file():
        all_files.append(p)

print(f"Total folders found: {len(all_dirs)}")
print(f"Total files found: {len(all_files)}")

if len(all_files) == 0:
    print("\nNo dataset files were found inside:", ROOT)
    print("This usually means Google Drive was not mounted correctly, or the dataset path is different.")
    print("\nAvailable folders under /content/drive after mounting:")
    try:
        shown = 0
        for p in Path("/content/drive").rglob("*"):
            if p.is_dir():
                print(" -", p)
                shown += 1
            if shown >= 30:
                break
    except Exception:
        pass
    raise FileNotFoundError(
        f"No files found under {ROOT}. Verify the path in Google Drive, then rerun from the first cell."
    )

inventory_rows = []
for fp in tqdm(all_files, desc="Building file inventory"):
    size_bytes, modified_time = safe_stat(fp)
    ext = fp.suffix.lower() if fp.suffix else "[no_extension]"
    mime_type, _ = mimetypes.guess_type(str(fp))
    rel_path = str(fp.relative_to(ROOT))
    nature = detect_file_nature(fp)
    inventory_rows.append({
        "top_folder": top_folder(fp),
        "component_guess": infer_component_from_path(rel_path),
        "relative_path": rel_path,
        "file_name": fp.name,
        "extension": ext,
        "mime_type": mime_type,
        "file_nature": nature,
        "size_bytes": size_bytes,
        "size_readable": human_size(size_bytes),
        "modified_time": modified_time,
    })

file_inventory = pd.DataFrame(inventory_rows)
file_inventory = file_inventory.sort_values(["top_folder", "relative_path"]).reset_index(drop=True)

total_size_bytes = file_inventory["size_bytes"].fillna(0).sum()
print(f"Total data size: {human_size(total_size_bytes)}")
display(file_inventory.head(20))

# -----------------------------
# 4. Core summary tables
# -----------------------------
section("2. Creating folder, file-type, and suspicious-file summaries")

folder_summary = (
    file_inventory.groupby("top_folder", as_index=False)
    .agg(
        component_guess=("component_guess", lambda x: ", ".join(sorted(set(x)))),
        total_files=("file_name", "count"),
        total_size_bytes=("size_bytes", "sum"),
        file_types=("extension", lambda x: ", ".join(sorted(set(x)))),
        suspicious_files=("file_nature", lambda x: sum(v in ["html_saved_as_data", "empty_or_unreadable", "json_or_api_response"] for v in x)),
        netcdf_files=("extension", lambda x: sum(v in [".nc", ".nc4"] for v in x)),
        csv_files=("extension", lambda x: sum(v == ".csv" for v in x)),
        excel_files=("extension", lambda x: sum(v in [".xlsx", ".xls"] for v in x)),
        parquet_files=("extension", lambda x: sum(v == ".parquet" for v in x)),
        geotiff_files=("extension", lambda x: sum(v in [".tif", ".tiff"] for v in x)),
        zip_files=("extension", lambda x: sum(v == ".zip" for v in x)),
    )
)
folder_summary["total_size_readable"] = folder_summary["total_size_bytes"].apply(human_size)
folder_summary = folder_summary.sort_values("total_size_bytes", ascending=False).reset_index(drop=True)

type_summary = (
    file_inventory.groupby(["extension", "file_nature"], as_index=False)
    .agg(file_count=("file_name", "count"), total_size_bytes=("size_bytes", "sum"))
)
type_summary["total_size_readable"] = type_summary["total_size_bytes"].apply(human_size)
type_summary = type_summary.sort_values("total_size_bytes", ascending=False).reset_index(drop=True)

suspicious_files = file_inventory[
    file_inventory["file_nature"].isin(["html_saved_as_data", "empty_or_unreadable", "json_or_api_response"])
    | (file_inventory["size_bytes"].fillna(0) < 1000)
].copy()

display(folder_summary)
display(type_summary)
print("Suspicious files detected:", len(suspicious_files))
display(suspicious_files.head(30))

# -----------------------------
# 5. NetCDF deep inspection
# -----------------------------
section("3. Inspecting NetCDF files: variables, dimensions, spatial and temporal coverage")

import xarray as xr

NETCDF_EXTS = [".nc", ".nc4"]
valid_netcdf_inventory = file_inventory[
    file_inventory["extension"].isin(NETCDF_EXTS)
    & file_inventory["file_nature"].isin(["valid_netcdf3", "valid_netcdf4_hdf5"])
].copy()

netcdf_rows = []

COORD_CANDIDATES = {
    "lat": ["lat", "latitude", "LATITUDE", "nav_lat", "y"],
    "lon": ["lon", "longitude", "LONGITUDE", "nav_lon", "x"],
    "time": ["time", "valid_time", "TIME", "JULD"],
    "depth": ["depth", "deptht", "lev", "pressure", "PRES", "PRES_ADJUSTED"],
}


def safe_array_minmax(ds, names, is_time=False):
    for name in names:
        if name in ds.coords or name in ds.variables:
            try:
                arr = ds[name].values
                if arr.size == 0:
                    continue
                if is_time:
                    t = pd.to_datetime(arr.ravel(), errors="coerce")
                    t = pd.Series(t).dropna()
                    if len(t) > 0:
                        return str(t.min()), str(t.max())
                    continue
                arr = np.asarray(arr).astype("float64").ravel()
                arr = arr[np.isfinite(arr)]
                if arr.size > 0:
                    return float(np.nanmin(arr)), float(np.nanmax(arr))
            except Exception:
                continue
    return None, None

for _, row in tqdm(valid_netcdf_inventory.iterrows(), total=len(valid_netcdf_inventory), desc="Inspecting NetCDF"):
    fp = ROOT / row["relative_path"]
    out = {
        "top_folder": row["top_folder"],
        "component_guess": row["component_guess"],
        "relative_path": row["relative_path"],
        "file_name": row["file_name"],
        "status": "ok",
        "error": "",
        "data_vars": "",
        "coords": "",
        "dims": "",
        "lat_min": np.nan,
        "lat_max": np.nan,
        "lon_min": np.nan,
        "lon_max": np.nan,
        "depth_min": np.nan,
        "depth_max": np.nan,
        "time_min": "",
        "time_max": "",
    }
    try:
        ds = xr.open_dataset(fp, decode_times=True, mask_and_scale=False)
        out["data_vars"] = safe_str(", ".join(list(ds.data_vars)))
        out["coords"] = safe_str(", ".join(list(ds.coords)))
        out["dims"] = safe_str(json.dumps({str(k): int(v) for k, v in ds.sizes.items()}))

        out["lat_min"], out["lat_max"] = safe_array_minmax(ds, COORD_CANDIDATES["lat"])
        out["lon_min"], out["lon_max"] = safe_array_minmax(ds, COORD_CANDIDATES["lon"])
        out["depth_min"], out["depth_max"] = safe_array_minmax(ds, COORD_CANDIDATES["depth"])
        out["time_min"], out["time_max"] = safe_array_minmax(ds, COORD_CANDIDATES["time"], is_time=True)
        ds.close()
    except Exception as e:
        out["status"] = "failed"
        out["error"] = safe_str(e, 5000)
    netcdf_rows.append(out)

netcdf_summary = pd.DataFrame(netcdf_rows)
print("NetCDF files inspected:", len(netcdf_summary))
display(netcdf_summary.head(30))

# -----------------------------
# 6. GeoTIFF/raster inspection
# -----------------------------
section("4. Inspecting GeoTIFF bathymetry files")

try:
    import rasterio
    raster_rows = []
    raster_inventory = file_inventory[file_inventory["extension"].isin([".tif", ".tiff"])].copy()
    for _, row in tqdm(raster_inventory.iterrows(), total=len(raster_inventory), desc="Inspecting GeoTIFF"):
        fp = ROOT / row["relative_path"]
        out = {
            "top_folder": row["top_folder"],
            "component_guess": row["component_guess"],
            "relative_path": row["relative_path"],
            "file_name": row["file_name"],
            "status": "ok",
            "error": "",
            "width": np.nan,
            "height": np.nan,
            "count": np.nan,
            "crs": "",
            "bounds": "",
            "nodata": "",
            "min_value_sample": np.nan,
            "max_value_sample": np.nan,
        }
        try:
            with rasterio.open(fp) as src:
                out["width"] = src.width
                out["height"] = src.height
                out["count"] = src.count
                out["crs"] = safe_str(src.crs)
                out["bounds"] = safe_str(src.bounds)
                out["nodata"] = safe_str(src.nodata)
                # Read a decimated sample for min/max
                sample = src.read(1, out_shape=(max(1, src.height // 20), max(1, src.width // 20))).astype("float32")
                if src.nodata is not None:
                    sample[sample == src.nodata] = np.nan
                out["min_value_sample"] = float(np.nanmin(sample))
                out["max_value_sample"] = float(np.nanmax(sample))
        except Exception as e:
            out["status"] = "failed"
            out["error"] = safe_str(e, 5000)
        raster_rows.append(out)
    raster_summary = pd.DataFrame(raster_rows)
except Exception as e:
    raster_summary = pd.DataFrame()
    print("Raster inspection skipped:", e)

display(raster_summary)

# -----------------------------
# 7. Tabular file inspection
# -----------------------------
section("5. Inspecting CSV, Excel, and Parquet files safely")

import pyarrow.parquet as pq

tabular_inventory = file_inventory[file_inventory["extension"].isin([".csv", ".xlsx", ".xls", ".parquet"])].copy()
tabular_rows = []

for _, row in tqdm(tabular_inventory.iterrows(), total=len(tabular_inventory), desc="Inspecting tabular"):
    fp = ROOT / row["relative_path"]
    ext = row["extension"]
    out = {
        "top_folder": row["top_folder"],
        "component_guess": row["component_guess"],
        "relative_path": row["relative_path"],
        "file_name": row["file_name"],
        "extension": ext,
        "status": "ok",
        "error": "",
        "columns": "",
        "column_count": np.nan,
        "row_count": np.nan,
        "row_count_method": "",
        "sheet_names": "",
        "detected_time_columns": "",
        "detected_lat_columns": "",
        "detected_lon_columns": "",
        "detected_depth_columns": "",
        "detected_key_columns": "",
    }
    try:
        columns = []
        if ext == ".csv":
            df_sample = pd.read_csv(fp, nrows=CSV_SAMPLE_ROWS, low_memory=False)
            columns = standardize_column_list(df_sample.columns.tolist())
            count, method = exact_csv_row_count(fp)
            out["row_count"] = count
            out["row_count_method"] = method
        elif ext in [".xlsx", ".xls"]:
            xls = pd.ExcelFile(fp)
            out["sheet_names"] = safe_str(", ".join(xls.sheet_names))
            df_sample = pd.read_excel(fp, sheet_name=xls.sheet_names[0], nrows=CSV_SAMPLE_ROWS)
            columns = standardize_column_list(df_sample.columns.tolist())
            out["row_count_method"] = "not_counted_excel"
        elif ext == ".parquet":
            pf = pq.ParquetFile(fp)
            columns = standardize_column_list(pf.schema.names)
            out["row_count"] = pf.metadata.num_rows if pf.metadata else np.nan
            out["row_count_method"] = "parquet_metadata"

        out["columns"] = safe_str(", ".join(columns))
        out["column_count"] = len(columns)
        cols_lower = [c.lower() for c in columns]
        out["detected_time_columns"] = safe_str(", ".join([c for c in columns if any(k in c.lower() for k in ["time", "date", "timestamp"])]))
        out["detected_lat_columns"] = safe_str(", ".join([c for c in columns if "lat" in c.lower()]))
        out["detected_lon_columns"] = safe_str(", ".join([c for c in columns if "lon" in c.lower()]))
        out["detected_depth_columns"] = safe_str(", ".join([c for c in columns if any(k in c.lower() for k in ["depth", "pres", "pressure"])]))
        out["detected_key_columns"] = safe_str(", ".join([c for c in columns if any(k in c.lower() for k in ["basin", "profile", "cyclone", "storm", "station", "platform"])]))
    except Exception as e:
        out["status"] = "failed"
        out["error"] = safe_str(e, 5000)
    tabular_rows.append(out)

tabular_summary = pd.DataFrame(tabular_rows)
print("Tabular files inspected:", len(tabular_summary))
display(tabular_summary.head(30))

# -----------------------------
# 8. Canonical file discovery
# -----------------------------
section("6. Discovering canonical project-ready dataset files")

# Component definitions use real expected variable groups, not only folder names.
COMPONENT_EXPECTATIONS = {
    "Argo_Core_Profile": {
        "required_keywords": ["time", "lat", "lon", "temp", "sal", "pres"],
        "preferred_path_keywords": ["argo", "standard", "profile", "parquet"],
    },
    "NOAA_OISST": {
        "required_keywords": ["time", "lat", "lon", "sst", "anom"],
        "preferred_path_keywords": ["oisst", "combined", "parquet"],
    },
    "MODIS_Chlorophyll": {
        "required_keywords": ["date", "lat", "lon", "chlor"],
        "preferred_path_keywords": ["modis", "chlorophyll", "025deg", "parquet"],
    },
    "Copernicus_Physics": {
        "required_keywords": ["current", "uo", "vo", "zos", "mlotst"],
        "preferred_path_keywords": ["physics", "current", "uo", "vo", "zos", "mlotst", "parquet"],
    },
    "Copernicus_Waves": {
        "required_keywords": ["wave", "height", "period", "energy"],
        "preferred_path_keywords": ["wave", "daily", "parquet"],
    },
    "ERA5_Atmosphere": {
        "required_keywords": ["date", "lat", "lon", "wind", "pressure"],
        "preferred_path_keywords": ["era5", "daily", "grid", "parquet"],
    },
    "IBTrACS_Cyclone": {
        "required_keywords": ["cyclone", "storm", "time", "lat", "lon", "wind", "pressure"],
        "preferred_path_keywords": ["ibtracs", "cyclone", "parquet"],
    },
    "GEBCO_Bathymetry": {
        "required_keywords": ["lat", "lon", "depth", "bathymetry"],
        "preferred_path_keywords": ["gebco", "bathymetry", "025deg", "parquet"],
    },
    "INCOIS_OMNI_RAMA": {
        "required_keywords": ["time", "lat", "lon"],
        "preferred_path_keywords": ["rama", "validation", "parquet"],
    },
    "BGC_Argo": {
        "required_keywords": ["oxygen", "chlor", "bbp", "nitrate"],
        "preferred_path_keywords": ["bgc", "sprof", "parquet"],
    },
}

# Build searchable schema text for every file.
schema_lookup = {}
for _, r in tabular_summary.iterrows():
    schema_lookup[r["relative_path"]] = lower_join([r.get("columns", ""), r.get("file_name", ""), r.get("relative_path", "")])
for _, r in netcdf_summary.iterrows():
    schema_lookup[r["relative_path"]] = lower_join([r.get("data_vars", ""), r.get("coords", ""), r.get("file_name", ""), r.get("relative_path", "")])
for _, r in raster_summary.iterrows() if len(raster_summary) > 0 else []:
    schema_lookup[r["relative_path"]] = lower_join([r.get("file_name", ""), r.get("relative_path", ""), "elevation depth bathymetry lat lon"])

canonical_rows = []
for component, cfg in COMPONENT_EXPECTATIONS.items():
    for rel, schema_text in schema_lookup.items():
        rel_lower = rel.lower()
        required_hits = sum(1 for k in cfg["required_keywords"] if k.lower() in schema_text or k.lower() in rel_lower)
        preferred_hits = sum(1 for k in cfg["preferred_path_keywords"] if k.lower() in rel_lower)
        ext = Path(rel).suffix.lower()
        ext_score = {".parquet": 5, ".csv": 4, ".nc": 3, ".nc4": 3, ".xlsx": 2, ".tif": 1, ".tiff": 1}.get(ext, 0)
        processed_score = 3 if any(k in rel_lower for k in ["processed", "daily", "summary", "025deg", "standard"]) else 0
        score = required_hits * 10 + preferred_hits * 4 + ext_score + processed_score
        if score > 0:
            canonical_rows.append({
                "component": component,
                "relative_path": rel,
                "extension": ext,
                "required_hits": required_hits,
                "required_total": len(cfg["required_keywords"]),
                "preferred_hits": preferred_hits,
                "score": score,
                "schema_preview": safe_str(schema_text, 1000),
            })

canonical_candidates = pd.DataFrame(canonical_rows)
if len(canonical_candidates) > 0:
    canonical_candidates = canonical_candidates.sort_values(["component", "score"], ascending=[True, False]).reset_index(drop=True)
else:
    canonical_candidates = pd.DataFrame(columns=["component", "relative_path", "extension", "score"])

best_canonical = (
    canonical_candidates.sort_values(["component", "score"], ascending=[True, False])
    .groupby("component", as_index=False)
    .head(5)
    .reset_index(drop=True)
)

display(best_canonical)

# -----------------------------
# 9. Strict readiness scoring
# -----------------------------
section("7. Strict readiness scoring for project implementation")

readiness_rows = []
for component, cfg in COMPONENT_EXPECTATIONS.items():
    comp_candidates = canonical_candidates[canonical_candidates["component"] == component].copy()
    if len(comp_candidates) == 0:
        readiness_rows.append({
            "component": component,
            "detected": False,
            "best_score": 0,
            "required_coverage_ratio": 0.0,
            "best_file": "",
            "status": "missing",
            "recommendation": "Download or regenerate this dataset component before full fusion.",
        })
        continue

    best = comp_candidates.sort_values("score", ascending=False).iloc[0]
    ratio = best["required_hits"] / max(best["required_total"], 1)
    detected = ratio >= 0.45 or best["score"] >= 25
    status = "ready" if ratio >= 0.75 else ("partial" if detected else "weak_or_misdetected")
    recommendation = "Ready for Stage 1 fusion audit."
    if component == "Copernicus_Physics" and status != "ready":
        recommendation = "Physics-current dataset may be missing or confused with wave data. Check for uo, vo, zos, and mlotst files."
    elif component == "BGC_Argo" and status != "ready":
        recommendation = "BGC Argo appears absent. Oxygen/chlorophyll profile classes can be approximated from MODIS + Argo core, but true OMZ validation needs BGC Sprof."
    elif status != "ready":
        recommendation = "Detected partially. Verify schema and choose a canonical processed Parquet/CSV file."

    readiness_rows.append({
        "component": component,
        "detected": bool(detected),
        "best_score": float(best["score"]),
        "required_coverage_ratio": round(float(ratio), 3),
        "best_file": best["relative_path"],
        "status": status,
        "recommendation": recommendation,
    })

strict_readiness = pd.DataFrame(readiness_rows)
display(strict_readiness)

# -----------------------------
# 10. Variable dictionary and ML fusion risk table
# -----------------------------
section("8. Creating variable dictionary and fusion-risk analysis")

variable_dictionary = pd.DataFrame([
    ["Argo_Core_Profile", "temperature_c", "Vertical temperature profile", "Core feature for stratification, thermal anomaly, and transfer learning"],
    ["Argo_Core_Profile", "salinity_psu", "Vertical salinity profile", "Freshwater intrusion, stratification, basin contrast"],
    ["Argo_Core_Profile", "pressure_dbar/depth_m", "Vertical coordinate", "Depth-standardized profile construction"],
    ["NOAA_OISST", "sst_celsius", "Sea surface temperature", "Thermal anomaly and cyclone surface forcing"],
    ["NOAA_OISST", "anom_celsius", "SST anomaly", "Thermal anomaly class and climate-risk context"],
    ["MODIS_Chlorophyll", "chlor_a_mean", "Surface chlorophyll-a", "HAB/productivity signal"],
    ["Copernicus_Physics", "uo/vo/current_speed", "Ocean current vectors", "Upwelling, transport, cyclone disturbance"],
    ["Copernicus_Physics", "zos/mlotst", "Sea surface height and mixed layer thickness", "Risk and stratification support"],
    ["Copernicus_Waves", "wave_height/period/direction", "Wave state", "Climate-risk and cyclone-wave disturbance"],
    ["ERA5_Atmosphere", "u10/v10/wind_speed", "Atmospheric wind forcing", "Cyclone and wind-driven upwelling"],
    ["ERA5_Atmosphere", "msl/sp", "Pressure fields", "Cyclone intensity and atmospheric forcing"],
    ["IBTrACS_Cyclone", "storm track/wind/pressure", "Cyclone event labels", "Cyclone-induced disturbance class"],
    ["GEBCO_Bathymetry", "seafloor_depth_m", "Static bathymetry", "Coastal/open/deep-ocean context"],
    ["INCOIS_OMNI_RAMA", "in-situ station variables", "Validation observations", "Independent validation and Indian institutional relevance"],
], columns=["component", "variable_group", "meaning", "project_use"])

def component_status(component):
    s = strict_readiness.loc[strict_readiness["component"] == component, "status"]
    return s.iloc[0] if len(s) else "missing"

fusion_risk_table = pd.DataFrame([
    ["Temporal alignment", "Daily/8-day/3-hourly/hourly mismatch", "High", "Convert to daily grid and use rolling windows for 8-day MODIS"],
    ["Spatial alignment", "Different grids: 0.083°, 0.2°, 0.25°, station/profile points", "High", "Create common 0.25° grid plus nearest-neighbour/profile interpolation"],
    ["MODIS cloud gaps", "Chlorophyll missing in cloudy periods", "Medium", "Use valid_pixel_count, 8-day composites, anomaly features"],
    ["Argo sparse profiles", "Irregular profile locations/times", "High", "Create depth-standardized profiles and daily basin/depth summaries"],
    ["Cyclone labels", "Cyclone effects extend beyond track point", "High", "Create distance-to-cyclone and time-window labels"],
    ["BGC Argo missing", "True oxygen/chlorophyll vertical BGC validation absent", "High", "Optional: add BGC Sprof later; current project can use MODIS+Argo+RAMA"],
    ["Copernicus physics availability", "May be missing if only waves were retained", "High", "Verify uo, vo, zos, mlotst files before full model"],
    ["Excel row limit", "Large grid tables exceed Excel", "Low", "Use Parquet for ML; Excel only for summaries"],
], columns=["risk_area", "issue", "severity", "mitigation"])

display(variable_dictionary)
display(fusion_risk_table)

# -----------------------------
# 11. Generate figures
# -----------------------------
section("9. Generating publication-style dataset audit figures")

plt.rcParams.update({"figure.dpi": 140, "font.size": 10})

figure_records = []

def save_current_fig(fig_name, title):
    out = FIG_DIR / fig_name
    plt.tight_layout()
    plt.savefig(out, bbox_inches="tight")
    plt.show()
    figure_records.append({"figure": fig_name, "title": title, "path": str(out)})

# Figure 1: Folder size
plt.figure(figsize=(12, 6))
plot_df = folder_summary.sort_values("total_size_bytes", ascending=True)
plt.barh(plot_df["top_folder"], plot_df["total_size_bytes"] / (1024 ** 3))
plt.xlabel("Size (GB)")
plt.title("Figure 1. Storage contribution by dataset folder")
save_current_fig("fig01_folder_storage_gb.png", "Storage contribution by dataset folder")

# Figure 2: File count by folder
plt.figure(figsize=(12, 6))
plot_df = folder_summary.sort_values("total_files", ascending=True)
plt.barh(plot_df["top_folder"], plot_df["total_files"])
plt.xlabel("Number of files")
plt.title("Figure 2. File count by dataset folder")
save_current_fig("fig02_file_count_by_folder.png", "File count by dataset folder")

# Figure 3: File count by extension
plt.figure(figsize=(10, 5))
ext_counts = file_inventory["extension"].value_counts().sort_values()
plt.barh(ext_counts.index, ext_counts.values)
plt.xlabel("File count")
plt.title("Figure 3. File count by extension")
save_current_fig("fig03_extension_file_count.png", "File count by extension")

# Figure 4: Total size by extension
plt.figure(figsize=(10, 5))
ext_size = file_inventory.groupby("extension")["size_bytes"].sum().sort_values()
plt.barh(ext_size.index, ext_size.values / (1024 ** 3))
plt.xlabel("Size (GB)")
plt.title("Figure 4. Storage size by file extension")
save_current_fig("fig04_extension_storage_gb.png", "Storage size by file extension")

# Figure 5: Suspicious files by folder
plt.figure(figsize=(12, 5))
plot_df = folder_summary.sort_values("suspicious_files", ascending=True)
plt.barh(plot_df["top_folder"], plot_df["suspicious_files"])
plt.xlabel("Suspicious file count")
plt.title("Figure 5. Suspicious or invalid files by folder")
save_current_fig("fig05_suspicious_files_by_folder.png", "Suspicious files by folder")

# Figure 6: NetCDF files by folder
plt.figure(figsize=(12, 5))
plot_df = folder_summary.sort_values("netcdf_files", ascending=True)
plt.barh(plot_df["top_folder"], plot_df["netcdf_files"])
plt.xlabel("NetCDF file count")
plt.title("Figure 6. Valid NetCDF availability by folder")
save_current_fig("fig06_netcdf_availability.png", "Valid NetCDF availability by folder")

# Figure 7: Tabular row counts by file, top 30
row_df = tabular_summary.dropna(subset=["row_count"]).copy()
row_df = row_df[row_df["row_count"] > 0].sort_values("row_count", ascending=False).head(30)
plt.figure(figsize=(12, 8))
if len(row_df) > 0:
    plt.barh(row_df["file_name"][::-1], row_df["row_count"][::-1])
    plt.xscale("log")
plt.xlabel("Rows (log scale)")
plt.title("Figure 7. Largest tabular datasets by row count")
save_current_fig("fig07_largest_tabular_row_counts.png", "Largest tabular datasets by row count")

# Figure 8: Strict readiness score
plt.figure(figsize=(11, 5))
plot_df = strict_readiness.sort_values("best_score", ascending=True)
plt.barh(plot_df["component"], plot_df["best_score"])
plt.xlabel("Strict readiness score")
plt.title("Figure 8. Dataset component readiness score")
save_current_fig("fig08_strict_readiness_score.png", "Dataset component readiness score")

# Figure 9: Required coverage ratio
plt.figure(figsize=(11, 5))
plot_df = strict_readiness.sort_values("required_coverage_ratio", ascending=True)
plt.barh(plot_df["component"], plot_df["required_coverage_ratio"])
plt.xlabel("Required variable coverage ratio")
plt.xlim(0, 1.05)
plt.title("Figure 9. Required variable coverage by component")
save_current_fig("fig09_required_variable_coverage.png", "Required variable coverage by component")

# Figure 10: NetCDF variable frequency
var_counter = Counter()
for text in netcdf_summary["data_vars"].fillna(""):
    for v in [x.strip() for x in text.split(",") if x.strip()]:
        var_counter[v] += 1
var_freq = pd.DataFrame(var_counter.items(), columns=["variable", "file_count"]).sort_values("file_count", ascending=False).head(30)
plt.figure(figsize=(12, 8))
if len(var_freq) > 0:
    plt.barh(var_freq["variable"][::-1], var_freq["file_count"][::-1])
plt.xlabel("Number of NetCDF files")
plt.title("Figure 10. Most frequent NetCDF variables")
save_current_fig("fig10_netcdf_variable_frequency.png", "Most frequent NetCDF variables")

# Figure 11: Spatial coverage rectangles from NetCDF files
plt.figure(figsize=(10, 6))
spatial_df = netcdf_summary.dropna(subset=["lat_min", "lat_max", "lon_min", "lon_max"]).copy()
for _, r in spatial_df.iterrows():
    try:
        x0, x1 = float(r["lon_min"]), float(r["lon_max"])
        y0, y1 = float(r["lat_min"]), float(r["lat_max"])
        if np.isfinite([x0, x1, y0, y1]).all():
            plt.plot([x0, x1, x1, x0, x0], [y0, y0, y1, y1, y0], alpha=0.25)
    except Exception:
        pass
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Figure 11. Spatial coverage envelopes of NetCDF datasets")
plt.grid(True, alpha=0.3)
save_current_fig("fig11_netcdf_spatial_coverage.png", "Spatial coverage envelopes of NetCDF datasets")

# Figure 12: Time coverage by NetCDF component
coverage_rows = []
for _, r in netcdf_summary.iterrows():
    if r.get("time_min") and r.get("time_max"):
        try:
            coverage_rows.append({
                "component": r["component_guess"],
                "file_name": r["file_name"],
                "start": pd.to_datetime(r["time_min"]),
                "end": pd.to_datetime(r["time_max"]),
            })
        except Exception:
            pass
coverage_df = pd.DataFrame(coverage_rows)
plt.figure(figsize=(12, 6))
if len(coverage_df) > 0:
    component_order = list(coverage_df["component"].dropna().unique())
    comp_to_y = {c: i for i, c in enumerate(component_order)}
    for _, r in coverage_df.iterrows():
        y = comp_to_y.get(r["component"], 0)
        plt.plot([r["start"], r["end"]], [y, y], linewidth=2, alpha=0.35)
    plt.yticks(range(len(component_order)), component_order)
plt.xlabel("Time")
plt.title("Figure 12. NetCDF temporal coverage by component")
plt.grid(True, axis="x", alpha=0.3)
save_current_fig("fig12_netcdf_temporal_coverage.png", "NetCDF temporal coverage by component")

# Figure 13: File size vs file count by folder
plt.figure(figsize=(9, 6))
plt.scatter(folder_summary["total_files"], folder_summary["total_size_bytes"] / (1024 ** 3))
for _, r in folder_summary.iterrows():
    plt.annotate(str(r["top_folder"])[:18], (r["total_files"], r["total_size_bytes"] / (1024 ** 3)), fontsize=7)
plt.xlabel("File count")
plt.ylabel("Total size (GB)")
plt.title("Figure 13. Dataset scale profile: file count versus storage")
plt.grid(True, alpha=0.3)
save_current_fig("fig13_file_count_vs_storage.png", "Dataset scale profile")

# Figure 14: Canonical candidate scores
plt.figure(figsize=(12, 6))
if len(best_canonical) > 0:
    top_best = best_canonical.sort_values("score", ascending=False).head(30)
    labels = [f"{r.component}: {Path(r.relative_path).name[:25]}" for r in top_best.itertuples()]
    plt.barh(labels[::-1], top_best["score"][::-1])
plt.xlabel("Candidate score")
plt.title("Figure 14. Highest-ranked canonical project files")
save_current_fig("fig14_canonical_candidate_scores.png", "Highest-ranked canonical project files")

figure_catalog = pd.DataFrame(figure_records)
display(figure_catalog)

# -----------------------------
# 12. Save tables and final report
# -----------------------------
section("10. Saving audit tables, Excel workbook, and Markdown report")

TABLES = {
    "01_file_inventory": file_inventory,
    "02_folder_summary": folder_summary,
    "03_file_type_summary": type_summary,
    "04_suspicious_files": suspicious_files,
    "05_netcdf_summary": netcdf_summary,
    "06_raster_summary": raster_summary,
    "07_tabular_summary": tabular_summary,
    "08_canonical_candidates": canonical_candidates,
    "09_best_canonical_files": best_canonical,
    "10_strict_readiness": strict_readiness,
    "11_variable_dictionary": variable_dictionary,
    "12_fusion_risk_table": fusion_risk_table,
    "13_figure_catalog": figure_catalog,
}

# Save CSV tables
for name, df in TABLES.items():
    out_csv = TABLE_DIR / f"{name}.csv"
    df.to_csv(out_csv, index=False)

# Save Excel workbook
excel_report = REPORT_DIR / "STAGE0_deep_ocean_dataset_audit_report.xlsx"
with pd.ExcelWriter(excel_report, engine="openpyxl") as writer:
    for sheet_name, df in TABLES.items():
        safe_df = df.copy()
        for col in safe_df.columns:
            if safe_df[col].dtype == "object":
                safe_df[col] = safe_df[col].apply(lambda x: safe_str(x, 32000))
        safe_df.to_excel(writer, sheet_name=sheet_name[:31], index=False)

# Save Markdown summary
md_report = REPORT_DIR / "STAGE0_deep_ocean_dataset_audit_report.md"
summary_lines = []
summary_lines.append("# Stage 0 Deep Ocean Dataset Audit Report\n")
summary_lines.append(f"Generated: `{datetime.now()}`\n")
summary_lines.append(f"Dataset root: `{ROOT}`\n")
summary_lines.append(f"Total folders: **{len(all_dirs)}**\n")
summary_lines.append(f"Total files: **{len(all_files)}**\n")
summary_lines.append(f"Total size: **{human_size(total_size_bytes)}**\n")
summary_lines.append("\n## Key Readiness Result\n")
summary_lines.append(strict_readiness.to_markdown(index=False))
summary_lines.append("\n\n## Folder Summary\n")
summary_lines.append(folder_summary.to_markdown(index=False))
summary_lines.append("\n\n## File Type Summary\n")
summary_lines.append(type_summary.to_markdown(index=False))
summary_lines.append("\n\n## Suspicious Files\n")
if len(suspicious_files) > 0:
    summary_lines.append(suspicious_files[["top_folder", "relative_path", "file_nature", "size_readable"]].head(50).to_markdown(index=False))
else:
    summary_lines.append("No suspicious files detected.\n")
summary_lines.append("\n\n## Best Canonical Files\n")
summary_lines.append(best_canonical[["component", "relative_path", "score", "required_hits", "required_total"]].head(50).to_markdown(index=False))
summary_lines.append("\n\n## Fusion Risk Table\n")
summary_lines.append(fusion_risk_table.to_markdown(index=False))
summary_lines.append("\n\n## Figure Catalog\n")
summary_lines.append(figure_catalog.to_markdown(index=False))
md_report.write_text("\n".join(summary_lines), encoding="utf-8")

print("Saved Excel report:", excel_report)
print("Saved Markdown report:", md_report)
print("Saved tables folder:", TABLE_DIR)
print("Saved figures folder:", FIG_DIR)

# -----------------------------
# 13. Create safe ZIP package
# -----------------------------
section("11. Zipping audit report for download")

# IMPORTANT:
# Do NOT create the ZIP inside REPORT_DIR while zipping REPORT_DIR.
# That causes the ZIP file to include itself repeatedly and may produce:
# OSError: [Errno 28] No space left on device
# RuntimeError: File size too large, try using force_zip64

import zipfile

# Remove any partial ZIP files accidentally created inside REPORT_DIR
for bad_zip in REPORT_DIR.glob("*.zip"):
    try:
        print("Removing old/partial internal ZIP:", bad_zip)
        bad_zip.unlink()
    except Exception as e:
        print("Could not remove:", bad_zip, e)

zip_path = ROOT / "STAGE0_deep_ocean_dataset_audit_report_LIGHT.zip"

if zip_path.exists():
    zip_path.unlink()

include_suffixes = {".xlsx", ".md", ".csv", ".png"}

files_to_zip = []
for fp in REPORT_DIR.rglob("*"):
    if fp.is_file() and fp.suffix.lower() in include_suffixes:
        files_to_zip.append(fp)

print("Files selected for ZIP:", len(files_to_zip))
print("ZIP output will be outside report folder:", zip_path)

with zipfile.ZipFile(zip_path, mode="w", compression=zipfile.ZIP_DEFLATED, allowZip64=True) as zf:
    for fp in files_to_zip:
        arcname = fp.relative_to(REPORT_DIR)
        zf.write(fp, arcname=str(arcname))

print("ZIP report created:", zip_path)
print("ZIP size:", human_size(zip_path.stat().st_size))

try:
    from google.colab import files
    files.download(str(zip_path))
except Exception as e:
    print("Automatic browser download skipped. You can manually download from:", zip_path)

# -----------------------------
# 14. Final console guidance
# -----------------------------
section("12. Audit completed — interpretation guide")

print("Use the following files for the next full project implementation step:")
print("1. STAGE0_deep_ocean_dataset_audit_report.xlsx")
print("2. STAGE0_deep_ocean_dataset_audit_report.md")
print("3. tables/09_best_canonical_files.csv")
print("4. tables/10_strict_readiness.csv")
print("5. figures/*.png")

print("\nImportant checks before Stage 1 model implementation:")
for _, r in strict_readiness.iterrows():
    print(f"- {r['component']}: {r['status']} | {r['recommendation']}")

print("\nStage 0 finished successfully at:", datetime.now())
```
