```python
import copernicusmarine
from pathlib import Path

DATASET_ID = "cmems_mod_glo_phy_my_0.083deg_P1D-m"

OUTPUT_DIR = Path("copernicusmarine_data/physics_reanalysis")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

BASINS = {
    "bob": {
        "name": "Bay_of_Bengal",
        "minimum_longitude": 80,
        "maximum_longitude": 98,
        "minimum_latitude": 5,
        "maximum_latitude": 22
    },
    "as": {
        "name": "Arabian_Sea",
        "minimum_longitude": 60,
        "maximum_longitude": 75,
        "minimum_latitude": 5,
        "maximum_latitude": 22
    }
}

YEARS = [2020, 2021, 2022, 2023, 2024]


def download_surface_currents(basin_key, year):
    basin = BASINS[basin_key]

    copernicusmarine.subset(
        dataset_id=DATASET_ID,
        variables=["uo", "vo"],
        minimum_longitude=basin["minimum_longitude"],
        maximum_longitude=basin["maximum_longitude"],
        minimum_latitude=basin["minimum_latitude"],
        maximum_latitude=basin["maximum_latitude"],
        start_datetime=f"{year}-01-01T00:00:00",
        end_datetime=f"{year}-12-31T23:59:59",
        minimum_depth=0,
        maximum_depth=1,
        output_filename=f"CMEMS_PHY_{basin_key}_surface_currents_uo_vo_{year}.nc",
        output_directory=str(OUTPUT_DIR),
        file_format="netcdf",
        netcdf_compression_level=4,
        skip_existing=True
    )


def download_surface_ocean_variables(basin_key, year):
    basin = BASINS[basin_key]

    copernicusmarine.subset(
        dataset_id=DATASET_ID,
        variables=["zos", "mlotst"],
        minimum_longitude=basin["minimum_longitude"],
        maximum_longitude=basin["maximum_longitude"],
        minimum_latitude=basin["minimum_latitude"],
        maximum_latitude=basin["maximum_latitude"],
        start_datetime=f"{year}-01-01T00:00:00",
        end_datetime=f"{year}-12-31T23:59:59",
        output_filename=f"CMEMS_PHY_{basin_key}_zos_mlotst_{year}.nc",
        output_directory=str(OUTPUT_DIR),
        file_format="netcdf",
        netcdf_compression_level=4,
        skip_existing=True
    )


for basin_key in BASINS:
    for year in YEARS:
        print(f"Downloading surface currents: {basin_key}, {year}")
        download_surface_currents(basin_key, year)

        print(f"Downloading zos and mlotst: {basin_key}, {year}")
        download_surface_ocean_variables(basin_key, year)

print("All required Copernicus physics subsets downloaded successfully.")
```

```python
import shutil
from pathlib import Path

source_folder = Path("copernicusmarine_data/physics_reanalysis")
zip_name = "copernicus_physics_reanalysis_bob_as_2020_2024"

shutil.make_archive(zip_name, 'zip', source_folder)

print("Zip created:", zip_name + ".zip")
```
